```python
def filter_doc_feature(line):
    docid = ''
    try: 
#         topic_json = json.loads(line.encode('utf-8'))
        topic_json = json.loads(line)
        if 'docid' in topic_json and len(topic_json['docid']) == 16 and topic_json['docid'][8:12].isdigit():
            docid = topic_json['docid'].encode('utf-8')
            yield (docid, topic_json)
        if 'docid' in topic_json and topic_json['docid'].startswith('V'):
            docid = topic_json['docid'].encode('utf-8')
            yield (docid, topic_json)
    except:
        pass
```

```python
import datetime
import time 

def filter_expire(line):
     
    docid,value = line
    today = datetime.datetime.now()
    last_month = today  +  datetime.timedelta(days= -31)
    yesterday = int(time.mktime(last_month.timetuple()))
    #yesterday = 1493568000
    is_return = 0
    if 'expireTime' in value and value['expireTime'] != "":
        timeArray = time.strptime(str(value['expireTime']), "%Y-%m-%d %H:%M:%S")
        expire_ts = int(time.mktime(timeArray))
        if expire_ts > yesterday:
            is_return = 1
    if 'expireTime2' in value and value['expireTime2'] != "":
        timeArray = time.strptime(str(value['expireTime2']), "%Y-%m-%d %H:%M:%S")
        expire_ts = int(time.mktime(timeArray))
        if expire_ts > yesterday:
            is_return = 1

    if is_return == 1:
        yield value
```

```python
def distinct_docid(doc_list):
    doc_dic = {}
    for item in doc_list:
        docid,value = item
        if docid not in doc_dic:
            doc_dic[docid] = 0
            yield item
```

```python
def parse_docid_feature(line):

    import mmhash
    TOPIC_VERSION = '0000 0000 0000 0000 0010 0001'
    VEDIO_TITLE_KW_V2_PROFILE_VERSION = '0011 0000 0000 0001 1100 0001'
    VEDIO_MANUALLY_CATEGORY_PROFILE_VERSION = '0011 0000 0000 0001 0010 0001'
    VEDIO_MANUALLY_CATEGORY1_PROFILE_VERSION = '0011 0000 0000 0001 1010 0001'
    VEDIO_EDIT_KW_PROFILE_VERSION =  '0011 0000 0000 0000 1100 0001'
    VEDIO_INTEREST_PROFILE_VERSION = '0011 0000 0000 0001 0100 0001'
    HEADLINE_TITLE_KW_PROFILE_VERSION = '0000 0000 0000 0001 1100 0001'
    HEADLINE_DOC_KW_PROFILE_VERSION =   '0000 0000 0000 0000 0101 0001'
    HEADLINE_EDIT_KW_PROFILE_VERSION =  '0000 0000 0000 0000 1100 0001'
    HEADLINE_INTEREST_PROFILE_VERSION = '0000 0000 0000 0001 0100 0001'
    HEADLINE_MANUALLY_CATEGORY_PROFILE_VERSION = '0000 0000 0000 0001 0010 0001'
    HEADLINE_LEVEL_CATEGORY_PROFILE_VERSION =    '0000 0000 0000 0001 1010 0001'
    HEADLINE_W2V_PROFILE_VERSION =  '0000 0000 0000 0000 0010 0001' 

    video_title_kw_v2_field = str(hex(int(VEDIO_TITLE_KW_V2_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    video_manually_category_field = str(hex(int(VEDIO_MANUALLY_CATEGORY_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    video_manually_category1_field =  str(hex(int(VEDIO_MANUALLY_CATEGORY1_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    video_edit_kw_field  = str(hex(int(VEDIO_EDIT_KW_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    video_interest_field = str(hex(int(VEDIO_INTEREST_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_title_kw_field = str(hex(int(HEADLINE_TITLE_KW_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_doc_kw_field = str(hex(int(HEADLINE_DOC_KW_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_edit_kw_field  = str(hex(int(HEADLINE_EDIT_KW_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_interest_field = str(hex(int(HEADLINE_INTEREST_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_manually_category_field = str(hex(int(HEADLINE_MANUALLY_CATEGORY_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_level_category_field = str(hex(int(HEADLINE_LEVEL_CATEGORY_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()
    headline_w2v_field = str(hex(int(HEADLINE_W2V_PROFILE_VERSION.replace(' ',""),2)))[2:].upper()

    def parse_headline_title_kw_v2_topic(feature_json):
        topic_list = []
        if 'title_kw_v2' in feature_json and feature_json['title_kw_v2'] != None and feature_json['title_kw_v2'] != "": 
            for item in feature_json['title_kw_v2']:
                topicid = mmhash.get_hash(item['label_word'].encode('utf-8'))
                score = item['label_score']
                if topicid != 0:
                    topic_list.append((topicid,score))
            if len(topic_list) > 5:
                topic_list = sorted(topic_list, key=lambda d: d[1], reverse=True)[0:5]
            output_list = []
            for item in topic_list:
                output_list.append(item[0])
            return output_list

    def parse_headline_doc_kw_topic(feature_json):
        topic_list = []
        if 'kw' in feature_json and feature_json['kw'] != None and feature_json['kw'] != "": 
            for item in feature_json['kw']:
                topicid = mmhash.get_hash(item['label_word'].encode('utf-8'))
                score = item['label_score']
                if topicid != 0:
                    topic_list.append((topicid,score))
            if len(topic_list) > 5:
                topic_list = sorted(topic_list, key=lambda d: d[1], reverse=True)[0:5]
            output_list = []
            for item in topic_list:
                output_list.append(item[0])
            return output_list

    def parse_headline_edit_kw_topic(feature_json):
        topic_list = []
        if 'dkeys' in feature_json and feature_json['dkeys'] != None and feature_json['dkeys'] != "": 
            for item in feature_json['dkeys'].split(','):
                topicid = mmhash.get_hash(item.encode('utf-8'))
                if topicid != 0:
                    topic_list.append(topicid)
            return topic_list

    def parse_headline_interest_topic(feature_json):
        topic_list = []
        if 'interests' in feature_json and feature_json['interests'] != None and feature_json['interests'] != "": 
            for item in feature_json['interests'].split(','):
                topicid = mmhash.get_hash(item.encode('utf-8').replace("POI",""))
                if topicid != 0:
                    topic_list.append(topicid)
            return topic_list

    def parse_headline_manually_category_topic(feature_json):
        topic_list = []
        if 'category' in feature_json and feature_json['category'] != None and feature_json['category'] != "": 
            word = feature_json['category'].split(',')[0]
            topicid = mmhash.get_hash(word.encode('utf-8'))
            if topicid != 0:
                topic_list.append(topicid)
            return topic_list


    def parse_headline_w2v_level_topic(feature_json):
        topic_list = []
        if 'topic_w2v' in feature_json and feature_json['topic_w2v'] != None and feature_json['topic_w2v'] != "": 
            for item in feature_json['topic_w2v']:
                if 'level_id' in item:
                    topicid = item['level_id']
                    score = item['label_score']
                    topic_list.append((topicid,score))
            topic_list = sorted(topic_list, key=lambda d: d[1], reverse=True)
            output_level = 0
            if len(topic_list) > 0:
                output_level = topic_list[0][0]/16/1024/1024
            output_list = []
            if output_level > 0:
                output_list.append(output_level)
            return output_list

    def parse_video_title_kw_v2_topic(feature_json):
        topic_list = []
        if 'title_kw_v2' in feature_json and feature_json['title_kw_v2'] != None and feature_json['title_kw_v2'] != "": 
            for item in feature_json['title_kw_v2']:
                topicid = mmhash.get_hash(item['label_word'].encode('utf-8'))
                if topicid == 0:
                    continue
                score = item['label_score']
                topic_list.append((topicid,score))
            if len(topic_list) > 5:
                topic_list = sorted(topic_list, key=lambda d: d[1], reverse=True)[0:5]
            output_list = []
            for item in topic_list:
                output_list.append(item[0])
            return output_list

    def parse_video_manually_category_topic(feature_json):
        topic_list = []
        if 'category' in feature_json and feature_json['category'] != None : 
            item = feature_json['category'].replace(',','/')
            if item.find('/') < 0:
                if len(item) > 6:
                    item = item.encode('utf-8')[0:6]
                    category = item + "/默认"
                else:
                    category = item.encode('utf-8') + "/默认"
            else:
                tuple = item.split('/')
                if len(tuple) >=2 and tuple[0] == tuple[1]:
                    category = tuple[0].encode('utf-8') + "/默认"
                else:
                    category = item.encode('utf-8')    
            topicid = mmhash.get_hash(category)
            if topicid != 0:
                topic_list.append(topicid)
                return topic_list

    def parse_video_manually_category1_topic(feature_json):
        topic_list = []
        if 'category' in feature_json and feature_json['category'] != None : 
            for item in feature_json['category'].split(','):
                topicid = mmhash.get_hash(item.encode('utf-8').split('/')[0])
                if topicid != 0:
                    topic_list.append(topicid)
            return topic_list

    def parse_video_edit_kw_topic(feature_json):
        topic_list = []
        if 'dkeys' in feature_json and feature_json['dkeys'] != None: 
            for item in feature_json['dkeys'].split(','):
                topicid = mmhash.get_hash(item.encode('utf-8'))
                if topicid != 0:
                    topic_list.append(topicid)
            return topic_list

    def parse_video_interest_topic(feature_json):
        topic_list = []
        if 'interests' in feature_json and feature_json['interests'] != None: 
            for item in feature_json['interests'].split(','):
                topicid = mmhash.get_hash(item.encode('utf-8').replace("POI",""))
                if topicid == 0:
                    continue
                topic_list.append(topicid)
            return topic_list


#     docid,(none,feature_json) = line
    docid, feature_json = line
    new_topic_dic = {}

    if docid.startswith(b'V') :

        video_title_kw_v2_topic = parse_video_title_kw_v2_topic(feature_json)
        if video_title_kw_v2_topic != None and len(video_title_kw_v2_topic) > 0:
            new_topic_dic[video_title_kw_v2_field] = video_title_kw_v2_topic
        
        video_edit_kw_topic = parse_video_edit_kw_topic(feature_json)
        if video_edit_kw_topic != None and len(video_edit_kw_topic) > 0:
            new_topic_dic[video_edit_kw_field] = video_edit_kw_topic
    
        video_interest_topic = parse_video_interest_topic(feature_json)
        if video_interest_topic != None and len(video_interest_topic) > 0:
            new_topic_dic[video_interest_field] = video_interest_topic

        video_manually_category_topic = parse_video_manually_category_topic(feature_json)
        if video_manually_category_topic != None and len(video_manually_category_topic) > 0:
            new_topic_dic[video_manually_category_field] = video_manually_category_topic 

        video_manually_category1_topic = parse_video_manually_category1_topic(feature_json)
        if video_manually_category1_topic != None and len(video_manually_category1_topic) > 0:
            new_topic_dic[video_manually_category1_field] = video_manually_category1_topic 
    else:

        headline_title_kw_topic = parse_headline_title_kw_v2_topic(feature_json)
        if headline_title_kw_topic != None and len(headline_title_kw_topic) > 0:
            new_topic_dic[headline_title_kw_field] = headline_title_kw_topic

        headline_doc_kw_topic = parse_headline_doc_kw_topic(feature_json)
        if headline_doc_kw_topic != None and len(headline_doc_kw_topic) > 0:
            new_topic_dic[headline_doc_kw_field] = headline_doc_kw_topic

        headline_edit_kw_topic = parse_headline_edit_kw_topic(feature_json)
        if headline_edit_kw_topic != None and len(headline_edit_kw_topic) > 0:
            new_topic_dic[headline_edit_kw_field] = headline_edit_kw_topic

        headline_interest_topic = parse_headline_interest_topic(feature_json)
        if headline_interest_topic != None and len(headline_interest_topic) > 0:
            new_topic_dic[headline_interest_field] = headline_interest_topic

        headline_manually_category_topic = parse_headline_manually_category_topic(feature_json)
        if headline_manually_category_topic != None and len(headline_manually_category_topic) > 0:
            new_topic_dic[headline_manually_category_field] = headline_manually_category_topic
    
        headline_w2v_topic = parse_headline_w2v_level_topic(feature_json)
        if headline_w2v_topic != None and len(headline_w2v_topic) > 0:
            new_topic_dic[headline_w2v_field] = headline_w2v_topic

    if len(new_topic_dic) > 0:
        return (docid, str(new_topic_dic))
```

```python
def merge_new_topic(old_dic, new_dic):
    if  isinstance(old_dic, str):
        old_dic = eval(old_dic)
    if  isinstance(new_dic, str):
        new_dic = eval(new_dic)
    for label_type in new_dic:
        if len(new_dic[label_type]) ==0:
            continue
        if label_type in old_dic and len(old_dic[label_type]) > 1000:
            continue
        if label_type not in old_dic:
            old_dic[label_type] = []
        old_dic[label_type] += new_dic[label_type]
    return old_dic
```
